### **1) Import libraries**

```python
import pandas as pd #data processing, CSV file
import numpy as np #linear algebra
import matplotlib.pyplot as plt #Data Visualization
```

### **2) Ingest data from CSV and Define new Datasets for ASEAN region**

```python
# Load the dataset "Violence Againts Women and Girls around the world"
violence = pd.read_csv('Violence_Against_Women.csv')
# Define a list of ASEAN Countries
asean = ['Brunei', 'Cambodia', 'Indonesia', 'Laos', 'Malaysia', 'Myanmar', 'Philippines', 'Singapore', 'Thailand', 'Vietnam']
# Filter the dataset to include only ASEAN countries
violence_asean = violence[violence['Country'].isin(asean)]
```

### **3)Manage different data types**

```python
#Convert the values in "Survey Year" into datetime objects
violence['Survey Year'] = pd.to_datetime(violence['Survey Year'])
```

### **4) Use your function for data analysis**

```python
#Define a function for data analysis
def analyze_data(df):
# Perform some analysis on the data
    analysis_result = df.groupby(['Country', 'Gender'])['Value'].mean()
    return analysis_result
# Use the function for data analysis
analysis_result = analyze_data(violence_asean)
analysis_result
```

### **5) Visualize Data to understand the common motives of violence**

```python
from wordcloud import WordCloud #library for word cloud visualizations
from wordcloud import STOPWORDS #filtered out text data
stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'maroon',
                      height =1000,
                      width = 1000
                     ).generate(str(violence_asean["Question"]))
plt.rcParams['figure.figsize'] = (10,10)
plt.axis("off")
plt.imshow(wordcloud)
plt.title("Motives of why this is happening ")
plt.show()
```

### **6) Data Wrangling by filtering only the case from those who has no educational background**

```python
violence_noeducation = violence_asean[violence_asean['Demographics Response'] == 'No education']
violence_noeducation.sample(3)
```
